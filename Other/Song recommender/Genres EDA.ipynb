```python
# Required Libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('genres.csv')
```

```python
# Drop rows where the song name is null
data = data.dropna(subset=['song_name'])

# Remove the columns 'Unnamed' and 'Title'
columns_to_drop = ['Unnamed: 0', 'title']
data = data.drop(columns=columns_to_drop)
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Assuming subset_data contains the mentioned features
features_to_scale = ['valence', 'energy', 'danceability', 'acousticness', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo']

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit and transform the specified features using the scaler
data_scaled = data.copy()  # Create a copy of the subset_data DataFrame
data_scaled[features_to_scale] = scaler.fit_transform(data_scaled[features_to_scale])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming subset_data_scaled contains the scaled data with 'valence' and 'energy' columns

# Plotting the points on the energy vs. valence plane
plt.figure(figsize=(8, 8))

# Plotting the points
plt.scatter(data_scaled['valence'], data_scaled['energy'], label='Data Points')

# Plotting x and y axes
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)


# Highlighting x and y axes
plt.text(0.95, 0.03, 'X-axis', color='blue', fontsize=12, ha='right')
plt.text(0.03, 0.95, 'Y-axis', color='blue', fontsize=12, ha='right')

# Labeling axes and title
plt.xlabel('Valence')
plt.ylabel('Energy')
plt.title('Energy vs. Valence')

# Displaying legend and grid
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
data_scaled["mood_vec"] = data_scaled.apply(lambda row: [row['valence'], row['energy']], axis=1)
data_scaled["mood_vec"].head()
```

```python
# Create a function to determine mood based on quadrant
def determine_mood(row):
    if row['mood_vec'][0] >= 0 and row['mood_vec'][1] >= 0:  # First quadrant
        return 'Happy'
    elif row['mood_vec'][0] < 0 and row['mood_vec'][1] >= 0:  # Second quadrant
        return 'Angry'
    elif row['mood_vec'][0] < 0 and row['mood_vec'][1] < 0:  # Third quadrant
        return 'Sad'
    else:  # Fourth quadrant
        return 'Calm'

# Apply the function to create the 'mood' column

data_scaled['mood'] = data_scaled.apply(determine_mood, axis=1)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Assuming 'data_scaled' is your DataFrame containing song information including 'song_name', 'mood', 'mood_vec' columns

# Define the list of available moods
available_moods = ['Sad', 'Happy', 'Angry', 'Calm']

# Ask user to choose a mood from available options
print("Choose a mood from the following options: Sad, Happy, Angry, Calm")
user_mood = input("Enter your mood: ")

# Check if the user input is one of the available moods
if user_mood not in available_moods:
    print("Invalid mood. Please select from available moods.")
else:
    # Filter the DataFrame based on the user's chosen mood
    filtered_df = data_scaled[data_scaled['mood'] == user_mood]

    # Check if there are songs available for the given mood
    if filtered_df.empty:
        print(f"No songs found for the mood '{user_mood}'.")
    else:
        # Assign unique IDs to the songs and create a mapping of IDs to song names
        filtered_df['song_id'] = range(1, len(filtered_df) + 1)
        id_to_song_mapping = filtered_df[['song_id', 'song_name']].set_index('song_id').to_dict()['song_name']

        # Display 5 random songs from the available songs in the mood
        print(f"Choose a song by entering its ID from the following 5 random {user_mood} songs:")
        random_songs = filtered_df.sample(n=min(5, len(filtered_df)))
        print(random_songs[['song_id', 'song_name']])

        # Ask the user to select a song by its ID
        try:
            selected_song_id = int(input("Enter the ID of the song: "))
            selected_song_name = id_to_song_mapping.get(selected_song_id)
            
            if selected_song_name:
                # Filter the DataFrame to get the mood_vec of the selected song
                mood_vec_selected = filtered_df.loc[filtered_df['song_id'] == selected_song_id, 'mood_vec'].values

                # Check if the selected song exists in the filtered DataFrame
                if len(mood_vec_selected) == 0:
                    print("Selected song not found.")
                else:
                    mood_vec_selected = np.array(mood_vec_selected[0])  # Convert to NumPy array

                    # Calculate cosine similarity of the selected song's mood_vec with other songs of the same mood
                    mood_vecs = np.vstack(filtered_df['mood_vec'])
                    similarity = cosine_similarity(mood_vecs, mood_vec_selected.reshape(1, -1))

                    # Exclude the selected song itself from the recommendation
                    similarity = np.squeeze(similarity)  # Remove single-dimensional entries
                    similar_indices = similarity.argsort()[::-1][1:]  # Sort indices by similarity, excluding the selected song
                    best_match_index = similar_indices[0]  # Get the index of the most similar song

                    # Get the best recommendation based on similarity
                    recommended_song = filtered_df.iloc[best_match_index]['song_name']
                    
                    print(f"Best recommendation for '{selected_song_name}' based on similarity in {user_mood} songs:")
                    print(recommended_song)
            else:
                print("Invalid song ID. Please enter a valid song ID.")
        except ValueError:
            print("Invalid input. Please enter a valid song ID.")
```
